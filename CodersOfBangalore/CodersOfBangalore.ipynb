```python
with open("finaldata.txt", encoding='utf-8') as f:   #earlier i had initial data(jo random thha ) now i am working on final data(eh shi data hai)
    data = f.read()
```

```python
chunks = data.split("\n\n") 
chunks = [c for c in chunks if len(c)>3]
chunks
```

```python
chunks[0]  #4th chunk show huaa 
```

```python
def parse_chunk(chunk):
    chunk = chunk.strip()
    sep_chunk = chunk.split('\n')

    username = sep_chunk[0]

    no_of_posts = int(sep_chunk[1].split(' post')[0].replace(',', ''))

    no_of_followers = float(
        sep_chunk[2].split(' follower')[0]
        .replace(',', '').replace('K', '').replace('M', '')
    )

    if 'K' in sep_chunk[2]:
        no_of_followers = int(no_of_followers * 1000)
    elif 'M' in sep_chunk[2]:
        no_of_followers = int(no_of_followers * 1_000_000)
    else:
        no_of_followers = int(no_of_followers)

    # Following 
    no_of_following = float(
        sep_chunk[3].split(' following')[0]
        .replace(',', '').replace('K', '').replace('M', '')
    )

    if 'K' in sep_chunk[3]:
        no_of_following = int(no_of_following * 1000)
    elif 'M' in sep_chunk[3]:
        no_of_following = int(no_of_following * 1_000_000)
    else:
        no_of_following = int(no_of_following)

    name = sep_chunk[4]

    if len(sep_chunk) > 5:
        type_of_page = sep_chunk[5]
        bio = "\n".join(sep_chunk[6:])
    else:
        type_of_page = "Unknown"
        bio = ""

    return {
        "username": username,
        "no_of_posts": no_of_posts,
        "no_of_followers": no_of_followers,
        "no_of_following": no_of_following,
        "name": name,
        "type_of_page": type_of_page,
        "bio": bio
    }


    # Page type and bio (if present)
    if len(sep_chunk) > 5:
        type_of_page = sep_chunk[5]
        bio = "\n".join(sep_chunk[6:])
    else:
        type_of_page = "Unknown"
        bio = ""

    return {
        "username": username,
        "no_of_posts": no_of_posts,
        "no_of_followers": no_of_followers,
        "no_of_following": no_of_following,
        "name": name,
        "type_of_page": type_of_page,
        "bio": bio
    }

```

```python
print(parse_chunk(chunks[0]))
```

```python
all_chunks=[]
for chunk in chunks:
    parsed_chunk=parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
print(all_chunks)
```

```python
import json
s=json.dumps(all_chunks,indent=4)
with open("data.json","w") as f:
    print(f.write(s))
```
